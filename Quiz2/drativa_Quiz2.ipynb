> **Note**: Keep in mind that the requested data type is a ***Dataframe*** object called **`averages_by_group`**.

### My Own Example 3

````
import matplotlib.pyplot as plt
from statsmodels.graphics.mosaicplot import mosaic

plt.rcParams["font.size"] = 16.0
mosaic(titanic.sort_values("Sex"), ["Sex", "Survived"])
plt.title("Relación entre género y supervivencia")
plt.show()
````

The above graph shows the number of survivors by gender on the Titanic. 0 represents deaths and 1 represents survivors. As explained in a previous graph, the number of female survivors was higher due to maritime tradition that reflected the values and social norms of the time. The officers and crew of the Titanic tried to follow this policy in an attempt to save as many lives as possible. However, although many women were rescued in the lifeboats, many of the boats left with empty spaces due to the lack of men willing to board the boats before all the women and children were safe. Many more lives could have been saved.
